```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/breast-cancer-dataset/breast-cancer.csv')
df.head()
```

```python
for i,j in enumerate(df.columns):
    print(i,j)
```

```python
df.info()
```

```python
print(df['diagnosis'].unique())
print(f"Total Two Unique Value Are There:",df['diagnosis'].nunique())
print(df['diagnosis'].value_counts())
```

```python
# M = 0 = 212_Count
# B = 1 = 357_Count
df['diagnosis'] = df['diagnosis'].replace({'M': 0,"B": 1})
```

```python
df.sample(3)
```

```python
New_columns = ['id' , 'radius_mean', 'texture_mean', 'perimeter_mean','area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
               'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean','radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
               'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se','fractal_dimension_se', 'radius_worst', 'texture_worst',
               'perimeter_worst', 'area_worst', 'smoothness_worst','compactness_worst', 'concavity_worst', 'concave points_worst','symmetry_worst',
               'fractal_dimension_worst','diagnosis']
df = df[New_columns]
# df = df.reindex(columns=New_columns)
```

```python
print(df.isna().sum())
sns.heatmap(df.isna())
```

```python
df.info()
```

```python
# Id is not Important For Prediction 
plt.figure(figsize=(14,9))
sns.heatmap(df.corr(),annot=True,annot_kws={"size":6})
plt.show()
```

```python
df.drop(columns='id',axis=1,inplace=True)
```

```python
for col in df.columns:
    if df[col].dtype != 'object' and col != 'diagnosis':
        plt.figure(figsize=(10,4))
        df.boxplot([col])
        plt.show()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# We are moving Without Trating OutLyers
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.12,stratify=y,random_state=41)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression()
model_lg = lg.fit(x_train,y_train)
model_lg
```

```python
y_pred = model_lg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve
con_mat = confusion_matrix(y_test,y_pred)
con_mat
```

```python
acc_scr = accuracy_score(y_test,y_pred)
acc_scr
```

```python
from sklearn import metrics
metrics.plot_roc_curve(model_lg,x_test,y_test)
plt.show()
```
